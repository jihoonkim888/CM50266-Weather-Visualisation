```python
## Task 1
# find csv files in folder...
def find_filename(path):
    import os
    return [f for f in os.listdir(path) if os.path.isfile(os.path.join(path, f))]

# Read and parse csv...
def read_csv(path, filename, codec):
    data = []
    with open(path+filename, 'r', encoding=codec) as f:
        header = [f.readline().strip('\n').split(',')]
#         print(header)
#         next(f)
        for line in f.readlines():
            line = line.strip('\n')
            data.append([v for v in line.split(',')])
            
    # deal with malinformed input...
    # reject line if any anomalies exist - no generation of exceptions...
    
    return header, np.array(data)

# compute min, max, mean and std for each component of weather data...
def feature(data):
    data = data.astype(np.float)
    minimum = np.amin(data) 
    maximum = np.amax(data)
    mean = np.mean(data)
    std = np.std(data)
    return minimum, maximum, mean, std

# A complete set of plots for identification of correlations between variables.
def corr(x, y, title, xlabel, ylabel, res=150):
    plt.figure(dpi=res)
    plt.scatter(x, y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

```python
path = './lab1-data/'
filenames = find_filename(path)
print('Minimum, maximum, mean and standard 1 deviation for each component:\n')
for filename in filenames:
    header, data = read_csv(path, filename, 'utf-8-sig')
    for i in range(1, data.shape[1]):
        print(filename.strip('.csv'), header[0][i], 'Min:', round(feature(data[:,i])[0], 2), 'Max:', 
              round(feature(data[:,i])[1], 2), 'Mean:', round(feature(data[:,i])[2], 2), 
              'Std:', round(feature(data[:,i])[3], 2))
```
